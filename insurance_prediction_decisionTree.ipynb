```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
path = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/TravelInsurancePrediction.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["TravelInsurance"] = df["TravelInsurance"].map({0: "Not Purchased", 1: "Purchased"})
```

```python
# explore the age column to see how age affects the purchase of an insurance policy:

import plotly.express as px
df = df
figure = px.histogram(df, x = "Age",
                      color = "TravelInsurance",
                      title= "Factors Affecting Purchase of Travel Insurance: Age")
figure.show()
```

```python
# explore how a person’s type of employment affects the purchase of an insurance policy:

figure = px.histogram(df, x = "Employment Type",
                      color = "TravelInsurance",
                      title= "Factors Affecting Purchase of Travel Insurance: Employment Type")
figure.show()
```

```python
df["GraduateOrNot"] = df["GraduateOrNot"].map({"No": 0, "Yes": 1})
df["FrequentFlyer"] = df["FrequentFlyer"].map({"No": 0, "Yes": 1})
df["EverTravelledAbroad"] = df["EverTravelledAbroad"].map({"No": 0, "Yes": 1})
df.head()
```

```python
x = np.array(df[["Age", "GraduateOrNot",
                   "AnnualIncome", "FamilyMembers",
                   "ChronicDiseases", "FrequentFlyer",
                   "EverTravelledAbroad"]])
y = np.array(df[["TravelInsurance"]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=72)
```

```python
model = DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
