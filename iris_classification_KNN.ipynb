```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
path = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/IRIS.csv'
```

```python
df = pd.read_csv(path)
```

```python
print(df.head())
```

```python
df.describe()
```

```python
print("Target Labels", df["species"].unique())
```

```python
# plot the df using a scatter plot which will plot the iris species according to
# the sepal length and sepal width:
import plotly.express as px
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
x = df.drop("species", axis=1)
y = df["species"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=72)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn
```

```python
knn.fit(x_train, y_train)
```

```python
# let’s input a set of measurements of the iris flower and use the model to
# predict the iris species
x_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(x_new)
print("Prediction: {}".format(prediction))
```
