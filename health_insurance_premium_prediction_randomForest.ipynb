```python
import numpy as np
import pandas as pd

import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
path = "/content/drive/MyDrive/Machine Learning projects with Python/datasets/Health_insurance.csv"
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.isnull().sum()
```

```python
# explore the distribution of people who smoke and who do not:
df = df
figure = px.histogram(df, x = "sex", color = "smoker", title= "Number of Smokers")
figure.show()
```

```python
df["sex"] = df["sex"].map({"female": 0, "male": 1})
df["smoker"] = df["smoker"].map({"no": 0, "yes": 1})
print(df.head())
```

```python
# explore the distribution of the regions where people are living according to
# the dataset:
pie = df["region"].value_counts()
regions = pie.index
population = pie.values
fig = px.pie(df, values=population, names=regions)
fig.show()
```

```python
x = np.array(df[["age", "sex", "bmi", "smoker"]])
y = np.array(df["charges"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=72)
```

```python
forest = RandomForestRegressor()
```

```python
forest
```

```python
forest.fit(x_train, y_train)
```

```python
ypred = forest.predict(x_test)
data = pd.DataFrame(data={"Predicted Premium Amount": ypred})
print(data.head())
```
