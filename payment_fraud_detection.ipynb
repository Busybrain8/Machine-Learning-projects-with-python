```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
path = "/content/drive/MyDrive/Machine Learning projects with Python/datasets/PS_20174392719_1491204439457_log.csv"
```

```python
data = pd.read_csv(path)
data.head(10)
```

```python
# check for null values
print(data.isnull().sum())
```

```python
# Exploring transaction type
print(data.type.value_counts())
```

```python
type = data["type"].value_counts()
```

```python
transactions = type.index
```

```python
quantity = type.values
```

```python
import plotly.express as px
figure = px.pie(data,
             values=quantity,
             names=transactions,hole = 0.5,
             title="Distribution of Transaction Type")
figure.show()
```

```python
# transform the categorical features into numerical.

data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2,
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
print(data.head())
```

**Payments Fraud Detection Model**

```python
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=42)
```

```python
model = DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
print(model.score(x_test, y_test))
```

```python
# prediction
#features = [type, amount, oldbalanceOrg, newbalanceOrig]
features = np.array([[1, 9000.60, 9000.60, 0.0]])
print(model.predict(features))
```
